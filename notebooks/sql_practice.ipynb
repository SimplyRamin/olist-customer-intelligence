```python
q("""
with revenue_per_order as (
    select 
    order_items.order_id,
    customers.customer_id,
    customers.customer_state,
    order_items.price + order_items.freight_value as revenue
    from order_items
    left join orders on order_items.order_id = orders.order_id
    left join customers on orders.customer_id = customers.customer_id
)
select
    customer_state,
    avg(revenue)
from revenue_per_order
group by customer_state
""")
```

```python
q("""
with revenue_per_order as (
    select 
    order_items.order_id,
    customers.customer_id,
    customers.customer_state,
    order_items.price + order_items.freight_value as revenue
    from order_items
    left join orders on order_items.order_id = orders.order_id
    left join customers on orders.customer_id = customers.customer_id
),
    avg_rev_per_ord_per_st as (
    select
        customer_state,
        avg(revenue) as avg_revenue
    from revenue_per_order
    group by customer_state
),
  
    avg_over_all_state as (
  select
    avg(avg_revenue) as overall_avg
  from avg_rev_per_ord_per_st
)

select
  *
from avg_rev_per_ord_per_st
cross join avg_over_all_state
where avg_rev_per_ord_per_st.avg_revenue > avg_over_all_state.overall_avg
""")
```

```python
q("""
with cte1 as (
    select
    customers.customer_unique_id,
    orders.order_purchase_timestamp,
    row_number() over(
        partition by customer_unique_id
        order by order_purchase_timestamp
    ) as order_rank,
    lag(order_purchase_timestamp) over(
        partition by customer_unique_id
        order by order_purchase_timestamp
    ) as previous_order_date
    from orders
    left join customers on customers.customer_id = orders.customer_id
  ),

    cte2 as (
    select
        *,
        datediff('day', previous_order_date, order_purchase_timestamp) as days_between
    from cte1
    where order_rank > 1
  )

select median(days_between) from cte2
""")
```

```python
q("""
WITH customer_stats AS (
    SELECT
        customers.customer_unique_id,
        SUM(order_items.price + order_items.freight_value) AS revenue,
        COUNT(DISTINCT orders.order_id) AS no_orders,
        AVG(order_reviews.review_score) AS avg_review_score
    FROM customers
    LEFT JOIN orders ON customers.customer_id = orders.customer_id
    LEFT JOIN order_items ON orders.order_id = order_items.order_id
    LEFT JOIN order_reviews ON orders.order_id = order_reviews.order_id
    WHERE orders.order_status = 'delivered'
    GROUP BY customers.customer_unique_id
)

SELECT *
FROM customer_stats
ORDER BY revenue DESC
LIMIT 10
""")
```

```python
q = ("""
WITH customer_order_years AS ( 
	SELECT 
		customers.customer_unique_id,
		MIN(order_purchase_timestamp) as first_order_date,
		MAX(order_purchase_timestamp) as last_order_date,
		COUNT(DISTINCT orders.order_id) as total_orders
	FROM customers
	LEFT JOIN orders ON customers.customer_id = orders.customer_id
	WHERE orders.order_status = 'delivered'
	GROUP BY customers.customer_unique_id
	HAVING total_orders > 1
)

SELECT
	*
FROM customer_order_years
WHERE YEAR(first_order_date) = 2017 AND YEAR(last_order_date) = 2018
""")
```

```python
q = ("""
WITH
  customer_orders AS (
    SELECT
      customers.customer_unique_id,
      order_purchase_timestamp,
      MIN(order_purchase_timestamp) OVER (
        PARTITION BY
          customers.customer_unique_id
      ) AS first_ever_order_date
    FROM
      customers
      LEFT JOIN orders ON orders.customer_id = customers.customer_id
  ),
  customer_return AS (
    SELECT
      *,
      CASE
        WHEN order_purchase_timestamp = first_ever_order_date THEN 'new'
        ELSE 'returning'
      END AS customer_situation
    FROM
      customer_orders
    WHERE
      YEAR (order_purchase_timestamp) = 2018
  )
SELECT
  SUM(
    CASE
      WHEN customer_situation = 'new' THEN 1
      ELSE 0
    END
  ) AS no_new,
  SUM(
    CASE
      WHEN customer_situation = 'returning' THEN 1
      ELSE 0
    END
  ) AS no_returning,
  MONTH (order_purchase_timestamp) AS month_order
FROM
  customer_return
GROUP BY
  month_order
""")
```
