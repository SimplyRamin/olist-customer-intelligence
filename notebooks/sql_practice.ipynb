```python
import duckdb
con = duckdb.connect('../data/olist.db')

def q(sql):
    return con.execute(sql).df()
```

day 1

```python
q("""
select * from orders
""")
```

```python
q("""
select 
  order_id, order_status, order_purchase_timestamp 
from orders 
where order_status = 'delivered'
""")
```

```python
q("""
select distinct
  order_status
from orders 
""")
```

```python
q("""
select
  order_id, order_status, order_purchase_timestamp
from orders
where order_status != 'delivered';
""")
```

```python
q("""
select
  order_status,
  count(*) as count  
from orders
group by order_status
""")
```

```python
q("""
select
  order_status,
  count(*) as count  
from orders
group by order_status
order by count DESC
""")
```

```python
q("""
select
  year(order_purchase_timestamp) as order_year,
  count(*) as count
from orders
where order_status = 'delivered'
group by 
  order_year
order by count DESC
""")
```

```python
q("""
select
  order_status,
  count(*) as count
from orders
group by order_status
having count > 1000
""")
```

```python
q("""
select
  customer_state,
  count(distinct customer_unique_id) as unique_customers
from orders
join customers on orders.customer_id = customers.customer_id
group by customer_state
order by unique_customers DESC
limit 10
""")
```

```python
q("""
select
  payment_type,
  avg(payment_value) as payment_avg
from order_payments
group by payment_type
""")
```

```python
q("""
select
  seller_id,
  count(*) as order_count
from order_items
group by seller_id
order by order_count DESC
limit 5
""")
```
